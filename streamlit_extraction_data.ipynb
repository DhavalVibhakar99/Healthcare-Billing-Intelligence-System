```python
import sqlite3
import pandas as pd

# Load full database
conn_local = sqlite3.connect('data/cms_medicare.db')
df = pd.read_sql_query("SELECT * FROM medicare_billing", conn_local)
conn_local.close()

# Create 10k row sample
sample = df.sample(n=10000, random_state=42)

# Save as deployment database
conn_deploy = sqlite3.connect('data/cms_medicare_deploy.db')
sample.to_sql('medicare_billing', conn_deploy, 
              if_exists='replace', index=False)
conn_deploy.close()

import os
size = os.path.getsize('data/cms_medicare_deploy.db')
print(f"Rows: {len(sample):,}")
print(f"File size: {size/1024/1024:.1f} MB")
```
